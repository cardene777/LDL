# Scrapy
- クローリングとスクレイピングを行うフレームワーク。
- ページの解析処理、保存処理など、各機能ごとにわかりやすく分割して記述できる。
- 基本的な処理はフレームワークが面倒を見てくれるので、スクレイピングに集中できる。
- どんなデータを取得して、どこに保存するのかという基本的な部分を書くだけでクローリングとスクレイピングができる。

### 操作手順
1. scrapyコマンドを使ってプロジェクトを作成
2. Spiderクラスを記述して、クロールとデータの取得処理を作成
    - 蜘蛛を小間使いとして使わしてさまざまな処理をさせるイメージ。
3. コマンドラインからscrapyを実行

```
scrapy startproject <プロジェクト名>

```
